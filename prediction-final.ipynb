## feature engineering

```python
# Reduce dataset to 30000 rows

# Parameters
input_file = '/kaggle/input/final-data/maindata.csv'  # Path to your CSV file
output_file = 'reduced_data.csv'  # Output dataset
join_column = 'join'  # Column with 'train'
sample_size = 30000  # Desired 30,000 rows from 'train'

# Function to sample rows
def sample_rows(file_path, join_col, sample_size):
    if not isinstance(file_path, str):
        raise TypeError(f"file_path must be a string, got {type(file_path)}")
    
    train_rows = []
    
    # Read the dataset in chunks for memory efficiency
    chunk_size = 100000  # Adjust based on memory
    try:
        for chunk in pd.read_csv(file_path, chunksize=chunk_size):
            # Filter rows for 'train'
            train_chunk = chunk[chunk[join_col] == 'train']
            
            # Append to list
            train_rows.append(train_chunk)
            
            # Check if we have enough samples
            train_count = sum(len(c) for c in train_rows)
            if train_count >= sample_size:
                break
        
        # Concatenate collected rows
        train_df = pd.concat(train_rows) if train_rows else pd.DataFrame()
        
        # Check available rows
        train_count = len(train_df)
        print(f"Available rows: {train_count} for 'train'")
        
        # Verify we have enough rows
        if train_count < sample_size:
            raise ValueError(f"Only {train_count} rows available for 'train', need {sample_size}")
        if train_count == 0:
            raise ValueError("No rows available for 'train'")
        
        # Sample exactly 30,000 rows
        final_df = train_df.sample(n=sample_size, random_state=42)
        
        return final_df
    
    except FileNotFoundError:
        raise FileNotFoundError(f"File not found: {file_path}")
    except KeyError:
        raise KeyError(f"Column '{join_col}' not found in the dataset")

# Execute sampling
try:
    data = sample_rows(input_file, join_column, sample_size)
    
    # Save the result
    data.to_csv(output_file, index=False)
    
    # Verify the counts
    print(f"Saved {len(data)} rows to {output_file}")
    print(data[join_column].value_counts())
except Exception as e:
    print(f"Error: {str(e)}")
```

```python
# === Drop unneeded columns ===
drop_cols = [
    'Unnamed: 0', 'Category_Health_Facility_UUID',
    'Month_Year', 'Month_Year_lat_lon', 'key',
    'Month_x', 'Month_y', 'Year_x', 'Year_y',
    'Transformed_Latitude_y', 'Transformed_Longitude_y', 'join']
df = data.drop(columns=[col for col in drop_cols if col in data.columns])
```

```python
# Define features and target
X = df.drop(columns=['Disease'])  # replace 'target' with your actual target column name
y = df['Disease']
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Label encode categorical columns in X
categorical_cols = X.select_dtypes(include='object').columns
le_dict = {}
for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    le_dict[col] = le  # Save encoder if needed for later

# Encode target
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)
y_onehot = to_categorical(y_encoded)
```

```python
from sklearn.preprocessing import StandardScaler

# Drop datetime columns
datetime_cols = X.select_dtypes(include='datetime64[ns]').columns
X = X.drop(columns=datetime_cols)


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# === Train/test split on full data (80/20) ===
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
)

# === Reshape for LSTM: [samples, timesteps, features] ===
X_train_lstm = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_lstm = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# === One-hot encode target ===
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
```

```python
# === Define LSTM model ===
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(64, input_shape=(1, X_train.shape[1]), return_sequences=False))
model.add(Dense(64, activation='relu'))
model.add(Dense(y_train_cat.shape[1], activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# === Training  model ===

from tensorflow.keras.callbacks import EarlyStopping

# Define early stopping
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,           # stop if val_loss doesn't improve after 5 epochs
    restore_best_weights=True
)

# Train the model with early stopping
history = model.fit(
    X_train_lstm, y_train_cat,
    validation_data=(X_test_lstm, y_test_cat),
    epochs=100,
    batch_size=64,
    callbacks=[early_stop]
)
```

```python
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import numpy as np

# === Predict on test set ===
y_pred_probs = model.predict(X_test_lstm)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test_cat, axis=1)

# === Accuracy & Classification Report ===
print("🔍 Accuracy:", accuracy_score(y_true, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_true, y_pred, target_names=le_target.classes_))
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# === 1. Train-test split ===
X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# === 2. Train LightGBM ===
lgbm = LGBMClassifier(n_estimators=100, random_state=42)
lgbm.fit(X_train, y_train)

# === 3. Predictions ===
y_pred = lgbm.predict(X_val)
y_pred_proba = lgbm.predict_proba(X_val)

# === 4. Accuracy & Classification Report ===
print("🔍 Accuracy:", accuracy_score(y_val, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_val, y_pred, target_names=le_target.classes_))

# === 5. AUC score (for multiclass, use average='macro') ===
auc_score = roc_auc_score(y_val, y_pred_proba, multi_class='ovr', average='macro')
print("🎯 Macro AUC Score:", auc_score)

# === 6. Plot ROC curves ===
fpr = {}
tpr = {}
for i in range(len(le_target.classes_)):
    fpr[i], tpr[i], _ = roc_curve(y_val == i, y_pred_proba[:, i])

plt.figure(figsize=(10, 6))
for i in range(len(le_target.classes_)):
    plt.plot(fpr[i], tpr[i], label=f"Class {le_target.classes_[i]}")
plt.plot([0, 1], [0, 1], 'k--', label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("📈 ROC Curves for LightGBM")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.utils.class_weight import compute_sample_weight

# === 1. Train-test split ===
X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# === 2. Compute sample weights for class balancing ===
sample_weights = compute_sample_weight(class_weight='balanced', y=y_train)

# === 3. Train LightGBM with sample weights ===
lgbm = LGBMClassifier(
    n_estimators=200,      # You can increase this further
    learning_rate=0.05,    # Lower learning rate helps deeper learning
    num_leaves=31,
    max_depth=-1,
    random_state=42
)
lgbm.fit(X_train, y_train, sample_weight=sample_weights)

# === 4. Predictions ===
y_pred = lgbm.predict(X_val)
y_pred_proba = lgbm.predict_proba(X_val)

# === 5. Accuracy & Classification Report ===
print("🔍 Accuracy:", accuracy_score(y_val, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_val, y_pred, target_names=le_target.classes_))

# === 6. AUC score (macro average for multiclass) ===
auc_score = roc_auc_score(y_val, y_pred_proba, multi_class='ovr', average='macro')
print("🎯 Macro AUC Score:", auc_score)

# === 7. Plot ROC curves ===
fpr = {}
tpr = {}
for i in range(len(le_target.classes_)):
    fpr[i], tpr[i], _ = roc_curve(y_val == i, y_pred_proba[:, i])

plt.figure(figsize=(10, 6))
for i in range(len(le_target.classes_)):
    plt.plot(fpr[i], tpr[i], label=f"{le_target.classes_[i]}")
plt.plot([0, 1], [0, 1], 'k--', label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("📈 ROC Curves for LightGBM")
plt.legend()
plt.grid(True)
plt.show()
```

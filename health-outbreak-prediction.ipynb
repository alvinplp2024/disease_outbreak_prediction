```python
## import data manipulation libraries
import pandas as pd
import numpy as np
```

```python
# Load datasets
train_df = pd.read_csv('/kaggle/input/health/Healthcare/outsmarting-outbreaks-challenge/Dataset/Train.csv')
toilets_df = pd.read_csv('/kaggle/input/health/Healthcare/outsmarting-outbreaks-challenge/Dataset/toilets.csv')
water_df = pd.read_csv('/kaggle/input/health/Healthcare/outsmarting-outbreaks-challenge/Dataset/water_sources.csv')
waste_df = pd.read_csv('/kaggle/input/health/Healthcare/outsmarting-outbreaks-challenge/Dataset/waste_management.csv')
test_df = pd.read_csv('/kaggle/input/health/Healthcare/outsmarting-outbreaks-challenge/Dataset/Test.csv')
```

```python
## inspect columns from the three datasets
print("waste columns\n")
print(waste_df.columns.tolist())
print("\nwater columns\n")
print(water_df.columns.tolist())
print("\ntoilets columns\n")
print(toilets_df.columns.tolist())
```

```python
## create a new column to make a key identification for stacking
waste_df['key'] = 'waste'
water_df['key'] = 'water'
toilets_df['key'] = 'toilets'

## create a key on text and train to stack the data
train_df['join'] = 'train'
test_df['join'] = 'test'
```

```python
## check column in a list
print(waste_df.columns.tolist())
```

```python
waste_df.head()
```

```python
# Stack the dataframes
stacked_df = pd.concat([waste_df, water_df, toilets_df], ignore_index=True)
stacked_df.head()
```

```python
# Stack the dataframes
stack_train_df = pd.concat([train_df, test_df], ignore_index=True)
stack_train_df.head()
```

## DATA PREPROCESSING AND CLEANING

```python
## CHECK DUPLICATES
stacked_df.duplicated().sum()
```

```python
##drop duplicates
data = stacked_df.drop_duplicates()
```

```python
## check duplicates
data.duplicated().sum()
```

```python
## duplicates for train
stack_train_df.duplicated().sum()
```

```python
# create a column to ue it to merge
stack_train_df['Month_Year'] = stack_train_df['Month'].astype(str) + '_' + stack_train_df['Year'].astype(str)
```

```python
stack_train_df.head()
```

```python
## merge columns to train dataset
merged = pd.merge(stack_train_df, data, on='Month_Year', how = 'left')
```
